```python
import numpy as np
from itertools import combinations
Edgefile = open(r"D:/Data Science Course/Text mining/جلسه هفتم/edge.csv", "w",encoding="utf-8")
Edgefile.write("Source,Target \n")
Nodefile = open(r"D:/Data Science Course/Text mining/جلسه هفتم/node.csv", "w",encoding="utf-8")
Nodefile.write("Id,Label \n")
dict=["\\u200c","\\u06cc","[","]","'","(",")"," "]
edge=[]
Edges=[]
allhashtags=[]
tempstr=""
for row in range(df.shape[0]):
    edge=()
    hashtags=df.iloc[row,3]
    edge=hashtags.split(',')
    newedge=edge
    for i in dict:
       newedge = [s.replace(i, "") for s in newedge]
    for i in newedge:
       allhashtags.append(i) 
    
    
    temp=combinations(newedge,2)
    for i in list(temp):
        tempstr=str(i)
        for i in dict:
            tempstr=(tempstr.replace(i,""))
       
        Edgefile.write("%s\n" % (tempstr,))
        
              
uniqueNode = list(set(allhashtags))


for i in uniqueNode:
    Nodefile.write("%s," % (i,))
    Nodefile.write("%s\n" % (i,))
   

Edgefile.close()
Nodefile.close()
```
